```python
import ast
import http.client
import requests
import time
from _config import api_key_jooble

def get_jobs(keyword, location, page):
    host = 'us.jooble.org'
    connection = http.client.HTTPConnection(host)
        
    #request headers
    body = '{"keywords": "%s", "location": "%s", "page": "%d"}' % (keyword, location, page)
    headers = {"Content-type": "application/x-www-form-urlencoded"}

    #json query
    connection.request('POST','/api/' + api_key_jooble, body, headers)
    response = connection.getresponse()
#     print(response.status, response.reason)
#     print(response.read())

    return ast.literal_eval(response.read().decode('utf-8'))
```

```python
# test variables to save time
# search_keyword = ['data scientist', 'data engineer']
# search_location = ['san diego', 'los angeles']
search_keyword = ['data scientist', 'data engineer', 'data analyst']
search_location = ['san diego', 'los angeles', 'san francisco', 'denver', 'austin', 'new york']
search_pages = 6

data = {}

for job in search_keyword:
    # new dictionary for city
    location_dict = {}
    
    for city in search_location:
        # list of pages
        page_list = []
        
        for page in range(search_pages):
            page_list.append(get_jobs(job, city, page)['jobs'])

        # load to dictionary with city as the key
        location_dict[(f'{city}')] = page_list
        
    # load to dictionary with job as the key
    data[(f'{job}')] = location_dict
```

```python
# the job listings get stored into 5 separate lists. this will merge them up one level.
for job in search_keyword:    
    for city in search_location:
        list1 = data[f'{job}'][f'{city}'][1]
        list2 = data[f'{job}'][f'{city}'][2]
        list3 = data[f'{job}'][f'{city}'][3]
        list4 = data[f'{job}'][f'{city}'][4]
        list5 = data[f'{job}'][f'{city}'][5]

        one_list = list1 + list2 + list3 + list4 + list5

        data[f'{job}'][f'{city}'] = one_list
```
